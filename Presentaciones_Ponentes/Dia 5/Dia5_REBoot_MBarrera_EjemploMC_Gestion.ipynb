**MMC en Gestión de Proyectos**

Para determinar la duración de un proyecto, muchas veces al momento de hacer el analisis de costo y riesgo, es necesario un poco más que la fecha esperada. 

Un producto es importado a Venezuela y de 10 veces, es estimado que tarda 80 dias en llegar a la ciudad de Caracas, y 30 en ser entregado a la USB. 

Para esto se puede utilizar el MMC. En este ejemplo realizaremos las siguientes tareas:

*   Definir la funcion PDF = beta
*   Generar números aleatorios dentro de esta distribución
*   Estimar el tiempo de dura un distribuidor en entregar un producto a la USB. 

```python
# -*- coding: utf-8 -*-
"""
Created on Thu Sep  2 10:10:22 2021

@author: mbarrera
"""
# Primero siempre importamos las librerias
import numpy as np
import math
import scipy.special as sc
import matplotlib.pyplot as plt
import random
import scipy.integrate as integrate
# Generamos listas vacias para almacenar los resultados de cada iteración
list1 = []
list2 = []
list3 = []
list4 = []
list5 = []
list6 = []
list7 = []
```

Segun los analistas de riesgo, ellos utilizan la PDF asociada a la funcion beta. Para esto, definiré los parametros de la funcíon funcion:

\begin{equation} 
PDF = B(a,b)
\end{equation}



```python
#Parametros inventados por Maria Teresa
n = 10
t_ciudad = 0.8 # de 0 a 1
a_ciudad = t_ciudad*n + 1 
b_ciudad = n - t_ciudad*n + 1
t_uni = 0.3 # de 0 a 1
a_uni = t_uni*n + 1
b_uni = n - t_uni*n + 1
```

Generamos números aleatorios dentro de estas distribuciones

```python
for i in range(0, 1000):
    t_c = random.betavariate(a_ciudad, b_ciudad)
    t_u = random.betavariate(a_uni, b_uni)
    t_t = (t_c + t_u )
    list1.append(t_c*100)
    list2.append(t_u*100)
    list3.append(t_t*100)
    list4.append(i)


fig, axs0 = plt.subplots(1, 3)
fig.set_figwidth(30)
axs0[0].hist(list1, 100, color='r')
axs0[1].hist(list2, 100, color='r')
axs0[2].scatter(list4, list1)
axs0[2].scatter(list4, list2)
axs0[2].axis([0, 10, 0, 100])
#plt.scatter(list4, list1)
#plt.scatter(list4, list2)
```
