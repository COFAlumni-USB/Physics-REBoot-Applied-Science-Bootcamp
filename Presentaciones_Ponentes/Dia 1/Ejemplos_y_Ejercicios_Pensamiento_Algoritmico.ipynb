# Matriz Diagonal

Vamos a crear una matriz diagonal

```python
import numpy as np
M=np.zeros([5,5])
x=np.arange(5)
y=np.arange(5)
```

```python
for i in x:
  for j in y:
    if i==j:
      M[i,j]=1
print(M)
```

# Nmero ms pequeo de Python

Vamos a intentar conseguir el nmero ms pequeo, o la incentidumbre de Python

```python
n=1.
m=0.5
while m != n:
    if m==0:
        break
    n=n/2
    m=m/2

print ("m=%e"%(m))
print ("n=%e"%(n))
```

# Propagacin de un incendio

Vamos a ver cmo se propaga un incendio bajo las siguientes condiciones: 


*   El bosque comienza sano
*   Un rbol se incendia
*   Por cada rbol incendiado, se incendias sus vecinos
*   Condiciones de borde peridicas



```python
import numpy as np

#Creamos bosque inicial
trees=np.ones([10,10])
print(trees)
```

```python
#Causamos primer quema del primer rbol

in_coor1=np.random.randint(0,10)
in_coor2=np.random.randint(0,10)
print("Coordenadas Iniciales=( %i , %i  )"%(in_coor1,in_coor2))
trees[in_coor1,in_coor2]=0
print(trees)
```

```python
#Matriz nueva a llenarse
Mnew=np.copy(trees)


#Condiciones bsicas para que cada vecino se incendie (Tomando en cuenta bordes)
for i in range(10):
  for j in range(10):
    if trees[np.mod(i+1,10),j]==0 or trees[np.mod(i-1,10),j]==0 or trees[i,np.mod(j+1,10)]==0 or trees[i,np.mod(j-1,10)]==0:
      Mnew[i,j]=0
trees=np.copy(Mnew)

print(trees)
```

Opcional: Propagar el incendio
