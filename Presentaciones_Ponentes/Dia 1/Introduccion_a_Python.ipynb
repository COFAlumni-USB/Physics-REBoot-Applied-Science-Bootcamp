Si este objeto a su vez tiene $j$ dimensiones, cada elemento tendra $j$ indices. Es importante tener esto en mente pues a traves de los ndices podemos manipular los objetos para realizar operaciones sobre grupos especficos de elementos.

De esta manera para seleccionar un objeto en especfico usaremos el nombre del objeto seguido de corchetes y los correspondientes ndices que describen la posicin del elemento deseado dentro del objeto
  
    Objeto[i,j,k,...] 

Estas propiedades de los ndices aplican para otros objetos en Python


```python
L_1 = [0,1,2,3]
print(L_1[0] )# Seleccionamos el primer elemento de L_1
L_2 = [[1,2,3],[4,5,6],[7,[8,9]]]
print(L_2)
print(L_2[0][1]) # Seleccionamos el segundo elemento del primer grupo de L_2
print(L_2[2][1][1]) # Seleccionamos el elemento con valor igual a 9


S = 'Hola Mundo'
print(S[1],S[3])
```

## Control de flujo


### `break`


Se utiliza dentro de un loop para interrumpir su ejecucin. En este caso el interprete ejecuta la seccin de cdigo posterior al loop.

```python
# Ejemplo:
D = [1,1,1,0,1,0,1,1]
N = 0
for item in D:
  if item == 1:
    N = N+1
  else:
    break
```

### `pass`

Se utiliza dentro de un loop para indicar que ningn cdigo ha de ejecutarse. Tpicamente se utiliza dentro de una seccin `if`, durante el desarrollo, cuando aun no esta bien definida la instruccin pero se desea conservar la estructura general del cdigo.


```python
# Ejemplo:
# Imprimimos en la consola los numeros del 1 al 10
D = list(range(1,11))
for d in D:
  if d%1 == 0:
    pass
  print(d)
```

Si no escribimos el `pass` tendremos un error.

###`continue`

Esta declaracin tiene un funcionamiento parecido a `pass`. Sin embargo, al momento de ser ejecutada, enva al intrprete al inicio del loop y prosigue a la siguiente interaccin, sin terminar de ejecutar el resto del loop de la iteracin previa.

```python
# Ejemplo:
# Imprimimos en la consola los numeros del 1 al 10. Sin embargo, al usar continue, nada sera impreso.
D = list(range(1,11))
for d in D:
  if d%1 == 0:
    continue
  print(d)
```

## Excepciones - `try/except`

`try:`

`except + "tipo de error":`

`...# Instruccin`

```python
# Ejemplo

D1 = [1,2,3,4,5,6,7]
D2 = [3,45,2,0,1,4,3]

for i in range(len(D1)):
  print(D1[i]/D2[i])
```

# Clases

Una clase es esencialmente un constructor de objetos. Estos objetos poseen ciertas propiedades en comn y `mtodos` o funciones intrnsecas que permiten realizar operaciones sobre ellos. 

Hemos visto hasta ahora varios ejemplos de clases sin saberlo:
  * Listas
  * Diccionarios
  * Variables


Dependiendo de la complejidad del proyecto, puede ser til usar clases ya que permiten conservar y ordenar de manera compacta toda la informacin relevante a la clase


`Clases` > `metodos` > `variables`

```python
class Vector_Op:
  '''
  Una clase que implementa varias funciones 
  '''
  def __init__(self, x, y):
    self.x = x
    self.y = y

  def Suma(self):
    if len(self.x) != len(self.y):
      print('Ambos vectores deben tener el mismo tamao')
    else:
      return [self.x[i]+self.y[i] for i in range(len(self.x))]
  
  def Diferencia(self):
    if len(self.x) != len(self.y):
      print('Ambos vectores deben tener el mismo tamao')
    else:
      return [self.x[i]-self.y[i] for i in range(len(self.x))]

  def Producto(self,type='inner'):
    if len(self.x) != len(self.y):
      print('Ambos vectores deben tener el mismo tamao')
    else:
      return [self.x[i]*self.y[i] for i in range(len(self.x))]

  def Producto_Interno(self,type='inner'):
    if len(self.x) != len(self.y):
      print('Ambos vectores deben tener el mismo tamao')
    else:
      return sum([self.x[i]*self.y[i] for i in range(len(self.x))])

  def Distancia_Quad(self):
    if len(self.x) != len(self.y):
      print('Ambos vectores deben tener el mismo tamao')
    else:
      return sum([(self.x[i]-self.y[i])**2 for i in range(len(self.x))])

v1 = [1,2,3]
v2 = [4,5,6]

S = Vector_Op(v1,v2)

print(S.Suma())
print(S.Diferencia())
print(S.Producto())
print(S.Producto_Interno())
print(S.Distancia_Quad())

```

#Librerias

Las librerias estan compuestas por mdulos que contienen a su vez definiciones de clases, funciones y otros objetos que son usados por la libreria.

Un ejemplo de **libreria** es `scipy` de la cual hablaremos mas tarde.

Un ejemplo de **mdulo** es `optimize`, el cual contiene varias rutinas para optimizacin de funciones. 

Para poder utilizar una libreria es necesario importarla dentro del cdigo que estemos usando. Siempre y cuando se encuentre instalada podemos cargar una libreria a traves del comando `import`:

E.g:

`import scipy`

Podemos importar incluso slo una seccin de la libreria, en caso de que slo nos interese realizar ciertas operaciones standalone:

E.g: 

`import scipy.optimize`

Tambien podemos usar alias, ya que cada vez que usemos una funcin de la libreria importada necesitamos indicar la libreria, el mdulo y la funcin:

E.g:

`Minimizer = scipy.optimize.minimize()`

Podemos facilitar esto usando un alias para la libreria y el mdulo. Esto lo hacemos al momento de import la libreria. Es importante sin embargo, no usar como alias una funcin predifinada en Python (E.g:`sum`), a fin de evitar problemas en el cdigo.

E.g:

`import scipy.optimize as optim`


`Minimizer = optim.minimize()`

Finalmente, es posible tambien importar slo una funcin o grupo de funciones de un mdulo en especifico, esto permite ahorrar espacio de memoria.

E.g:
