```python
import numpy as np
from scipy.integrate import odeint, ode
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
```

```python
theta0 = np.pi/2
omega0 = 0
m = 0.2
l = 10
g = 9.8

inicio = 0
final = 120
Num = 120

def dx(x, t, m, g, l): #La función se llamará dx

  theta, omega = x[0], x[1]

  dtheta = omega
  domega = -m*g*np.sin(theta)/l

  return [dtheta, domega]

x0 = [theta0, omega0] #condiciones iniciales
tspan = np.linspace(inicio, final, Num)#Dominio del tiempo.

results = solve_ivp(lambda t, x :dx(x, t, m, g, l), [tspan[0], tspan[-1]], x0, t_eval=tspan, rtol=1e-8)#Se emplea la función solve_ivp

x = odeint(dx, x0, tspan, args=(m,l,g)) #Se emplea la función odeint.
theta, omega = x.T
```

```python
#Graficando cada función en diferentes diagramas se obtiene:

fig = plt.figure()
ax1 = fig.add_subplot(1,2,1)
ax1.plot(results.t, theta) #Entrada 0 del vector x
ax1.set_ylabel('Posición Angular',fontweight='bold')
ax1.set_xlabel('Tiempo (s)',fontweight='bold')
ax1.grid()
ax2 = fig.add_subplot(1,2,2)
ax2.plot(results.t, omega) #Entrada 1 del vector x
ax2.set_ylabel('Velocidad Angular', fontweight='bold')
ax2.set_xlabel('Tiempo (s)',fontweight='bold')
ax2.grid()
plt.tight_layout()
```

```python
#En otro diagrama:

fig = plt.figure(facecolor='w')
ax = fig.add_subplot(111, facecolor='#dddddd', axisbelow=True)
ax.plot(tspan, theta, 'b', alpha=0.5, lw=2, label='Posición angular') #Variable S vs t

ax.set_xlabel('Tiempo (s)') #titulo del eje X
ax.set_ylabel('Posición angular') #titulo eje y
ax.set_ylim(-2,2) #min z max del eje Y
ax.yaxis.set_tick_params(length=0)
ax.xaxis.set_tick_params(length=0)
ax.grid(b=True, which='major', c='w', lw=2, ls='-')

legend = ax.legend()
legend.get_frame().set_alpha(0.5)

for spine in ('top', 'right', 'bottom', 'left'): 
    ax.spines[spine].set_visible(False)
    
plt.show()

fig = plt.figure(facecolor='w')
ax1 = fig.add_subplot(111, facecolor='#dddddd', axisbelow=True)
ax1.plot(tspan, omega, 'r', alpha=0.5, lw=2, label='Velocidad angular') # Variable I vs t

ax1.set_xlabel('Tiempo (s)') #titulo del eje X
ax1.set_ylabel('Velocidad angular') #titulo eje y
ax1.set_ylim(-1,1) #min z max del eje Y
ax1.yaxis.set_tick_params(length=0)
ax1.xaxis.set_tick_params(length=0)
ax1.grid(b=True, which='major', c='w', lw=2, ls='-')

legend = ax.legend()
legend.get_frame().set_alpha(0.5)

for spine in ('top', 'right', 'bottom', 'left'): 
    ax.spines[spine].set_visible(False)
    
plt.show()
```
