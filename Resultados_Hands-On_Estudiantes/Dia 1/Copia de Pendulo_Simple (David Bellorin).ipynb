#***El Pndulo Simple***

A continuacin se presenta la resolucin de las ecuaciones diferenciales asociadas al sistema fsico del pndulo simple o pndulo ideal.

El pndulo simple consta de una cuerda de longitud **l** (con masa despreciable) fijo en el extremo superior y con una masa puntual **m** (de dimensin despreciable) en el extremo inferior:

PenduloSimple.png

En el sistema hay dos fuerzas involucradas, la tensin **T** y el peso **mg**. 

Para escribir las ecuaciones de movimiento es necesario primero, descomponer las fuerzas de modo de que sean paralelas o perpendiculares al movimiento. En este caso el movimiento esta definido a lo largo de la circunferencia y el eje perpendicular corresponde a la direccin de la cuerda. 

Haciendo uso del diagrama de cuerpo libre presentado anteriormente, y la segunda ley de Newton, escribimos la ecuacin correspondiente a la aceleracin tangencial:

$F_{Tang}= -mgsen(\theta ) = ma_{Tang}$

donde  $a_{Tang} = l.\ddot{\theta}$

Sustituyendo la segunda ecucacin dentro de la primera obtenemos:

$ -gsen(\theta ) = l.\ddot{\theta}$

Con esta informacin, podemos excribir las ecuaciones de movimiento para definir el sistema, donde:

$ \dot{\theta}= \frac{d\theta}{dt} = \omega $

$ \ddot{\theta}= \frac{d\omega}{dt} = -mg.sen(\theta)/l $

En esta ocasin, nuestra tarea es:

* Crear nuestra propia implementacin en Python del *Pendulo Simple* usando la ecuacin diferencial presentada anteriormente.

* Ilustrar graficamente la evolucin del sistema en funcin del tiempo




Empezemos por cargar las librerias y paquetes que usaremos:

```python
import math 
import numpy as np #NumPy paquete de computo cientifico, permite hacer arrays y tambien tiene constantes y funciones
from scipy.integrate import  odeint #Integracion numerica de un sistema de ecuaciones ordinarias con condiciones iniciales Solve_ivp
from matplotlib import pyplot as plt
from matplotlib import animation as animation #Paquete de graficos
```

**Solve_ivp** requiere como input una funcin que consiste en las ecuaciones diferenciales a ser resueltas en forma de vector:

El primer paso consiste en definir las variables que van a ser determinadas:

Luego se define la funcin de sus derivadas, $\dot{X(t)} = f(t,X)$ donde **t** es la variable independiente y **X** el vector que definimos anteriormente.

Finalmente, se indican los valores iniciales, de igual manera en forma vectorial:

La funcin se define:

```python
# definir ecucacion de movimiento del pendulo como funcion
    
g = 9.8  # [m/s^2] 
l = 10 #[m]
m = 1 #[Kg]
Start = 0
Total_s = 10
Num = 300
t = np.linspace(Start, Total_s, Num)

def f(X, t, g, l ,m):
    theta = X[0]
    omega = X[1]
# Posicin angular theta, entrada 0 del vector X

# Velocidad Angular, entrada 1 del vector X

    return np.array([omega, -(m*g*math.sin(theta)/l)]) #(derivada de X[0], derivada de X[1]), (Velocidad Angular, Aceleracion)

X0 = [np.pi/2, 0] #Condiciones Iniciales en forma de vector [ , ]
results = odeint(f, X0, t, args=(g,l,m)) #solve_ivp(f, (0,11) , X0,rtol=1e-8) #funcion, intervalo de tiempo (t0, tf), vector condiciones iniciales, sensitividad?

[theta, omega] = results.T
```

Finalmente graficamos la solucin:


```python
fig = plt.figure()
ax1 = fig.add_subplot(1,2,1)
ax1.plot(t, theta) #Entrada 0 del vector X
ax1.set_ylabel('Posicin Angular',fontweight='bold')
ax1.set_xlabel('Tiempo (s)',fontweight='bold')
ax1.grid()
ax2 = fig.add_subplot(1,2,2)
ax2.plot(t, omega) #Entrada 1 del vector X
ax2.set_ylabel('Velocidad Angular', fontweight='bold')
ax2.set_xlabel('Tiempo (s)',fontweight='bold')
ax2.grid()
plt.tight_layout()
```
