```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp as sol
import math



# Se definen las constantes del sistema

g = 9.8 
l = 20
m = 5


# Se define la ecuación de movimiento del sistema

def f(t,X):
    x_ang = X[0] #posición angular
    v_ang = X[1] #velocidad angular
    dx_ang = v_ang #derivada de la posición angular
    dv_ang = -(m*g/l)*math.sin(x_ang) #derivada de la velocidad angular
    return np.array([dx_ang, dv_ang]) #devuelve derivada de posición y velocidad angular



# Se establecen los tiempos iniciales y finales

t_span = np.array([0,20])

# Se definen las condiciones iniciales en forma vectorial

C0 = [np.pi/2,0]



# Se resuelve el problema

solv = sol(f, t_span, C0, rtol=1e-8)



# Se grafica la solución

fig = plt.figure()
ax1 = fig.add_subplot(1,2,1)
ax1.plot(solv.t, solv.y[0])
ax1.set_ylabel('Posición Angular', fontweight='bold')
ax1.set_xlabel('Tiempo (s)', fontweight='bold')
ax1.grid()

ax2 = fig.add_subplot(1,2,2)
ax2.plot(solv.t, solv.y[1])
ax2.set_ylabel('Velocidad Angular', fontweight='bold')
ax2.set_xlabel('Tiempo (s)', fontweight='bold')
ax2.grid()
plt.tight_layout()

```
