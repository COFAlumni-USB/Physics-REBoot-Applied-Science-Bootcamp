#***El Juego de la Vida***


El 'Juego de la Vida' es un modelo matemtico y computacional para un sistema dinmico que evoluciona en pasos discretos. Concebido por John Horton Conway, el juego consiste en un array bidimensional infinito vaco sobre el cual algunas celdas se inicializan con **'vida'**, es decir algun nmero, tipicamente **1**.

El juego tiene un numero de reglas que se pueden condensar de la siguiente manera:


* Cualquier clula con **2** o **3** vecinos **'vivos'** sobrevive.
* Cualquier clula muerta con exactamente **3** vecinos **'vivos'** cobra vida.
* El resto de las clulas **'vivas'** mueren en la siguiente generacin. De manera similar, las clulas muertas permanecen muertas.


A lo largo del tiempo, nuevas variantes fueron creadas al introducir, remover o cambiar las reglas. Actualmente, existen varias implementaciones eficientes de este juego en distintos lenguajes y plataformas como **Golly**, **Mirek's Cellebration** y **Xlife**.

Describimos entonces la regin local de cada celda ***$i$*** 
La regin consiste en las clulas del entorno, es decir contando desde **(i-1, j-1)** hasta **(i+1, j+1)**.

Para esto definimos una funcin la cual producira un sub array con las celulas dentro de la region.

```python
# Definir una funcion con la region entorno a la celda i,j
def region_local(Grid,i,j):
    region = Grid[i-1:i+2,j-1:j+2]
    return region     
```

En el siguiente paso, basado en las reglas descritas en el principio, definimos la logica del programa que decidira si una clula vive o muere:

```python
# Definir funcion logica celda
def logica_celda(array):
    valor_celda = array[1,1]
    if valor_celda == 1:  
        N_vecinos_vivos = array.sum() - 1
    else:
        N_vecinos_vivos = array.sum()
    
    if N_vecinos_vivos >= 2 and N_vecinos_vivos <=3 and valor_celda == 1: 
        return array
    elif valor_celda == 1:
        array[1,1] = 0
        return array
    elif valor_celda == 0 and N_vecinos_vivos == 3:
        array[1,1] = 1
        return array
    else:
        return array
```
