**Ejercicio 2:**

La masa de una partícula tiene una distribución normalizada dada por la función:

$m(x) = a e^{-a x}$

donde x es una variable, $a=1/3$, y el dominio de la funcion es $x>0$.

¿Cuál es la probabilidad de que la partícula tenga una masa mayor que 5?

```python
from scipy.integrate import quad  #para  integrar
import math
import numpy as np


#definimos la funcion
a=1/3

def m(x):
  return a*math.exp(-a*x)

#hallamos la probabilidad intergando de 5 a infinito
P,err= quad(m,5,np.inf)

print('la probabilidad es', P)
```

**Ejercicio 3:**

Una distribución Gaussiana centrada en $\mu$ y con desviación estándar $\sigma$ está definida como:

f(x) = $\frac{1}{\sigma \sqrt{2 \pi}} e^{\frac{-1}{2}(\frac{x-\mu}{\sigma} )^2}$


Comprueba numéricamente que el valor esperado de esta distribución es $\mu$ y que su varianza es $\sigma ^2$.

Sugerencia:

El valor esperado $E[x]$ de una variable x con distribución de probabilidad dada por $f(x)$ se define como:

$E[x] = \int x f(x) dx$

Mientras que la varianza $V[x]$ se puede calcular como:
$V[x] = E[x^2] - (E[x])^2 $


```python
from scipy.integrate import quad
import math
import numpy as np

mu= 50
s= 10

print('parametros de la Distribucion Gaussiana son mu=%f y s= %f'  %(mu,s))
print('s^2=', s**2)

#definimos la distribucion gaussiana de media mu y desviacion estandar s
def Gauss(x):
  return (math.exp(-0.5*((x-mu)/s)**2)/(s*math.sqrt(2*math.pi)))


#definimos x*Gauss
def G1(x):
  return Gauss(x)*x

def G2(x):
  return Gauss(x)*x*x

E, errE = quad(G1,-np.inf,np.inf)

print('Cálculo numérico')
print('el valor esperado es ' , E)

#varianza
E2,errE2= quad(G2,-np.inf,np.inf)

S= E2-(E**2)

print('la varianza es ' , S)

```
